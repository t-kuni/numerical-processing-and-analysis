### ğŸŸ  å‡¦ç†ã®æµã‚Œ

1. æœ€åˆã®é … ( x ) ã‚’åˆæœŸå€¤ã¨ã—ã¦ `p` ã«ä»£å…¥ã€‚
   `c` ã«ã‚‚åŒã˜å€¤ã‚’å…¥ã‚Œã‚‹ã€‚
2. `k = 1` ã‹ã‚‰ `n-1` ã¾ã§ãƒ«ãƒ¼ãƒ—ã—ã¦æ¬¡ã®é …ã‚’é †ã«æ±‚ã‚ã‚‹ã€‚
   $$
   c = c \times \frac{-x^2}{(2k+1)(2k)}
   $$
   ã“ã‚Œã¯å‰ã®é …ã‹ã‚‰æ¬¡ã®é …ã‚’ä½œã‚‹ãŸã‚ã®å¼ã€‚
3. æ–°ã—ãæ±‚ã‚ãŸé … `c` ã‚’ `p` ã«åŠ ç®—ã—ã¦ã„ãã€‚

å‰ã®é …ã®è¨ˆç®—çµæœã‚’æµç”¨ã—ã¦ **è¨ˆç®—åŠ¹ç‡ã‚’é«˜ã‚ã¦ã„ã‚‹**

```python
def sin_maclaurin(x, n):
    p = x
    c = p
    for k in range(1, n):
        c = c * (-x**2) / ((2*k+1)*(2*k))
        p = p + c
    return p

# å‘¼ã³å‡ºã—ã‚µãƒ³ãƒ—ãƒ«
x = 1.0
n = 5
print(sin_maclaurin(x, n))
```
